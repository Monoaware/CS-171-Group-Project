#### **Loss & Accuracy:**

#### **Analysis:**

Removing the randomization factor was a mistake because not only did the model not train on every single class, but it also didn't get evaluated on every single class. Since the partial training set was still very large compared to the partial validation set, we suspect that the model was memorizing the classes during the training phase, resulting in very obvious overfitting. Once again, the correct strategy would have been to apply stratified random sampling.
